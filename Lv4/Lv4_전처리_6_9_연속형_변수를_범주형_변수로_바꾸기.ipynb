```python
# ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∞è Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
import pandas as pd

train = pd.read_csv('data/train.csv')
```

```python
# train Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏
train.head()
```

```python
# Î∞©Î≤ï1. ÏàòÏπò Î≤îÏúÑ Íµ¨Í∞ÑÏùÑ ÏßÅÏ†ë ÏßÄÏ†ïÌï¥ Î†àÏù¥Î∏îÎßÅ ÌïòÍ∏∞

# train Îç∞Ïù¥ÌÑ∞Ïùò pH Î≥ÄÏàòÎ•º Íµ¨Í∞ÑÏù¥ 4Í∞úÏù∏ Î≤îÏ£ºÌòï Î≥ÄÏàòÎ°ú Î≥ÄÌôò
# pH < 1  -> lowest
# 1<= pH < 2 -> low
# 2 <= pH < 3-> normal
# 3 <= pH -> high
 
def func(x):
  if x < 3: return 'lowest'
  elif x < 3.3: return 'low'
  elif x < 3.5: return 'normal'
  else: return 'high'

train['pH'] = train['pH'].apply(lambda x : func(x))
```
